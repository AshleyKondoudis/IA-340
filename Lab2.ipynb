## Group & Visualize the data

```python
quintile_by_population = df.groupby('Quintile')['Population'].sum() 
quintile_by_population.head()
```

```python
condition_totalpop_by_population = df.groupby('Condition_TotalPop')['Population'].sum() 
condition_totalpop_by_population.head()
```

### Visualization 1

```python
import matplotlib.pyplot as plt
df.plot.scatter(x='Population', y='Condition_TotalPop')
plt.xlabel('Population')
plt.ylabel('Condition_TotalPop')
plt.title('Distribution of COVID-19 in King County')
```

### Visualization 2

```python
condition_totalpop_by_population.hist()
plt.xlabel('Population')
plt.ylabel('Number of people with COVID-19 in the Population')
plt.title('Distribution of COVID-19 in King County')
```

#### Markdown Notes

Visualization 1: In this visualization, there is a somewhat positive correlation between population and the number of people who get COVID-19. When there is a larger population, there is a higher chance of getting COVID-19.
